```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('us_debt_31_05.csv')
dfc = df.copy()

dfc = dfc[['Record Date', 'Total Public Debt Outstanding']]
dfc = dfc.rename(columns={'Record Date': 'date', 'Total Public Debt Outstanding': 'debt'})

dfc['date'] = pd.to_datetime(dfc['date'])

dfc['year'] = dfc['date'].dt.year
last_day_of_year = dfc.groupby('year').apply(lambda x: x.loc[x['date'].idxmax()])

# Reset the index to get a clean DataFrame
last_day_of_year = last_day_of_year.reset_index(drop=True)

display(last_day_of_year)
```

```python
sns.set_style("whitegrid")

# Create the time series plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='debt', data=last_day_of_year)

# Add title and labels
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')

# Show the plot
plt.show()
```
